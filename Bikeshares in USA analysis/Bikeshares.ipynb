```python
import time
import pandas as pd
import numpy as np

```

```python
CITY_DATA = { 'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
              'washington': 'washington.csv' }
months = [ "january","february","march","april","may","june","all"]
days = ["saturday","sunday","monday","tuesday","wednesday","thursday","friday","all"]
              

def get_filters():
    
  
    city=input("please enter the name of city , chicago or new york city or washington : ").lower().strip()
    while city not in CITY_DATA :
        print("you have entered a wrong name , please check the letters you have entered ")
        city=input("please enter the name of city : chicago or new york city or washington : ").lower().strip()
        
    month=input("please enter a month to filter on , january or february or march or april or may or june or all : ").lower().strip()
    while month not in months :
        print("you have entered a wrong name , please check the letters you have entered ")
        month=input("please enter a month to filter by , january or february or march or april or may or june or all : ").lower().strip()
        
    day=input("please enter a day to filter on , saturday or sunday or monday or tuesday or wednesday or thursday or friday or all : ").lower().strip()
    while day not in days :
        print("you have entered a wrong name , please check the letters you have entered ")
        day=input("please enter a day to filter by ,saturday or sunday or monday or tuesday or wednesday or thursday or friday or all : ").lower().strip()
        
     
    print('Hello! Let\'s explore some US bikeshare data!')
    print('-'*40)
    return city, month, day


```

```python
def load_data(city, month, day):
   
    df=pd.read_csv(CITY_DATA[city])
    df["Start Time"]=pd.to_datetime(df["Start Time"])
    df["End Time"]=pd.to_datetime(df["End Time"])
    df["month"]=df["Start Time"].dt.month
    df["day"]=df["Start Time"].dt.strftime("%A")
    
    if month !="all":
        month=months.index(month)+1
        df=df[df["month"]==month]
        
    if day != 'all':
        df = df[df['day'] == day.title()]    
             
    return df    
        
```

```python
def time_stats(df):
    start_time = time.time()
    
    MostCommonMonth_Num=df['month'].mode()[0]
    MostCommonMonth=months[MostCommonMonth_Num-1]
    
    MostCommonDay=df["day"].mode()[0]
    
    df["hour"]=df["Start Time"].dt.hour
    MostCommonHour=df["hour"].mode()[0]
    
    print("the most common month "+MostCommonMonth+"\nthe most common day of the week "+ MostCommonDay+" \nthe most common hour "+str(MostCommonHour))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def station_stats(df):
    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()
    
    MostCommonStartStation=df["Start Station"].mode()[0]
    print("most common start station : " ,MostCommonStartStation)
    
    MostCommonEndStation=df["End Station"].mode()[0]
    print("most common end station : ",MostCommonEndStation)
    
    df["combination"]=df["Start Station"]+df["End Station"]
    MostFrequentCombination=df["combination"].mode()[0]
    print("most frequent combination of start station and end station : ",MostFrequentCombination)
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def trip_duration_stats(df):
    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # TO DO: display total travel time
    TotalTravelTime=df["hour"].sum()
    print("total travel time (hrs) : ", TotalTravelTime)

    # TO DO: display mean travel time
    MeanTravelTime=df["hour"].mean()
    print("mean travel time (hrs) : ", MeanTravelTime)

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def user_stats(df):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()
    df["Birth Year"]=pd.to_numeric(df["Birth Year"])
    UserTypes=df["User Type"].value_counts()
    print("counts of user types : ",UserTypes)
    
    Gender=df["Gender"].value_counts()
    print("counts of gender : ",Gender)


    EarliestYearOfBirth=df["Birth Year"].min()
    MostRecentYearOfBirth=df["Birth Year"].max()
    MostCommonYearOfBirth=df["Birth Year"].mode()[0]
    
    print("earliest year of bitrh : ",EarliestYearOfBirth,"\nmost recent year of birth : ",MostRecentYearOfBirth,"\nmost common year of birth : ",MostCommonYearOfBirth)
    
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
```

```python
def display_the_data(df):
    
    #Display contents of the CSV file as requested by the user.

    start_loc = 0
    end_loc = 5

    display = input("Do You Want To Display The Raw Data?: ").lower().strip()

    if display == 'yes':
        while end_loc <= df.shape[0] - 1:

            print(df[start_loc:end_loc])
            start_loc += 5
            end_loc += 5

            end_display = input("Would you like to continue?: ").lower().strip()
            if end_display == 'no':
                break
```

```python
def main():
    while True:
        city, month, day = get_filters()
        df = load_data(city, month, day)

        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)
        display_the_data(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
	main()
```
